```python
import numpy as np
import pandas as pd

import holoviews as hv
hv.extension('bokeh')
```

```python
df = pd.read_csv('./master.csv')
df = df[[
 'material',
 'stress_indicator',
 'sample',
 'file_tag',
 'bends',
 'primary_position',
 'frequency',
 'temperature',
 'pressure',
 'bend_to_break',
 'break_location',
 'prim_sec_amp',
 'prim_sec_phi',
 'prim_rec_amp',
 'prim_rec_phi',
 'sec_rec_amp',
 'sec_rec_phi',
 'sec_harm_db',
 'rec_harm_db'
]]
df.head()
```

```python
list(df.columns)
```

```python
%opts Curve [logx=False, logy=False, width=400, height=300]
%opts Curve [show_legend=True, show_grid=True, tools=['hover']]
%opts Scatter (size=4)

def doit(df, sample):
    df = df[df['sample'] == sample]
    material = df.material.iloc[0]
    stress = df.stress_indicator.iloc[0]
    breaking_loc = df.break_location.iloc[0]
    cols = [
        'prim_sec_amp',
        'sec_harm_db',
    ]

    ctot = None
    for col in cols:
        c = None
        for pos in sorted(df.primary_position.unique()):
            dfx = df[df.primary_position == pos]
            c_this = hv.Curve((dfx.bends, dfx[col]), label=f'{pos}', kdims=[f'Bends "{material}-{stress}-{breaking_loc}" {sample}'], vdims=[col])
            c_this *= hv.Scatter(c_this)
            if c is None:
                c = c_this
            else:
                c = c * c_this
        if ctot is None:
            ctot = c
        else:
            ctot += c
    ctot.cols(2)
    
    return ctot
for samp in df['sample'].unique():
    display(doit(df, samp))
```
